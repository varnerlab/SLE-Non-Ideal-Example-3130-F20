### Solid Liquid Equilibrium (SLE) Case 2: Non-ideal Liquid + Immiscible Solid

The fugacity matching condition for solid liquid equlibrium is given by:

$$
x_{i}\gamma_{i} = z_{i}\Gamma_{i}\psi_{i}\qquad{i=1,2,\dots,\mathcal{M}}
$$

where $x_{i}$ denotes the mol fraction of component $i$ in the liquid, $\gamma_{i}$ denotes the liquid phase activity coefficient, $z_{i}$ denotes the mol fraction of component $i$ in the solid, $\Gamma_{i}$ denotes the activity coefficient for component $i$ in the solid and $\psi_{i}$ denotes the temperature dependence of ratio of reference fugacities. The quantity $\psi_{i}$ is given by:

$$
\ln\psi_{i} = -\frac{\Delta{H_{i}^{ls}}}{R}\left(\frac{1}{T} - \frac{1}{T_{m_{i}}}\right)
$$

where $\Delta{H_{i}^{ls}}$ denotes the heat of fusion for pure component $i$, $T$ denotes the system temperature, $R$ denotes the ideal gas constant, and $T_{m_{i}}$ denotes the melting temperature for pure component $i$. 

#### Experimental validation: Ideal liquid + Ideal Solid

When we have ideal liquid, and an ideal solid, the activity coefficients for component $i$ in the liquid ($\gamma_{i}$) and solid ($\Gamma_{i}$) are unity. Thus, our design equation reduces to:

$$
x_{i}=z_{i}\psi_{i}\qquad{i=1,2,\dots,\mathcal{M}}
$$

Given the melting temperature, and the heat of fusion for each component, we can construct a Txz phase diagram for the different phases our mixture can exist in (at least in the binary case). Let's check this out with a binary ($\mathcal{M}$ = 2) example. We are going to pull some data from a recent 2019 study [Luo et. al., Ind. Eng. Chem. Res. 2019, 58, 10018âˆ’10030](https://pubs.acs.org/doi/full/10.1021/acs.iecr.9b00921). Let's simulate a glutaric acid(1)/acetic acid (2) mixture and compare with experimental data.

#### Experimental validation: Non ideal liquid + Ideal Solid (immiscible solid)

When we have a non-ideal liquid, and an ideal single component solid, the activity coefficients for component $i$ in the solid ($\Gamma_{i}$) are unity. Thus, our design equation reduces to:

$$
x_{i}\gamma_{i}=\psi_{i}\qquad{i=1,2,\dots,\mathcal{M}}
$$

where $\gamma_{i}$ can be estimated using a Gibbs excess energy model. [Luo et. al., Ind. Eng. Chem. Res. 2019, 58, 10018âˆ’10030](https://pubs.acs.org/doi/full/10.1021/acs.iecr.9b00921) use the [Non-random two liquid (NRTL) model of Prausnitz and coworkers](https://en.wikipedia.org/wiki/Non-random_two-liquid_model). Let's use the NRTL model to estimate the liquid phase activity coefficients, and compare our results with experimental data. However, as it turns out the parameters published on the Luo et al study are incorrect (don't believe everything you read!). We'll (re)estimate the parameters from experimental estimates of $\gamma_{i}$ by minimizing the squared difference between the activity coefficients estimated from data versus those calculated using the NRTL model:

$$
\min_{k} \sum_{i=1}^{\mathcal{E}}\left(\gamma_{i}-\gamma_{exp,i}\right)^{2}
$$

using the [Broydenâ€“Fletcherâ€“Goldfarbâ€“Shanno algorithm (BFGS)](https://en.wikipedia.org/wiki/Broydenâ€“Fletcherâ€“Goldfarbâ€“Shanno_algorithm) in the [Optim.jl](https://github.com/JuliaNLSolvers/Optim.jl) package, where $k$ denotes a vector of unknown parameters ($\alpha$,$a_{12}$, $a_{21}$, $b_{12}$ and $b_{21}$). 

```julia
# compute the *experimentally estimated* gamma values -
(number_of_temperature_steps,number_of_cols) = size(experimental_data_array)
experimental_gamma_array = Array{Float64,2}(undef,number_of_temperature_steps,2)

# estimate the gamma from the data and our psi models -
for row_index = 1:number_of_temperature_steps
    
    # get á´ª values -
    á´ª1 = psi_array[row_index,1]
    á´ª2 = psi_array[row_index,2]
    
    # get x values -
    x1 = experimental_data_array[row_index,2]
    x2 = 1 - x1 # binary
    
    # esimate gamma -> from our matching condition
    ð›„1 = (1/x1)*(á´ª1)
    ð›„2 = (1/x2)*(á´ª2)
    
    # package -
    experimental_gamma_array[row_index,1] =  ð›„1
    experimental_gamma_array[row_index,2] =  ð›„2
end
```

```julia
# compute the gamma values using the NRTL model -
# first (re)estimate the NRTL parameters - published parameters don't seem to fit the data???

# setup model parameters -> Table 6 in Luo et al 2019 will be the initial guess ...
pVec = [
    1.25 2.16 -3.10 -985.40 1339.70
]

# setup the calculation -
xinitial = pVec
OF(p) = obj_function_NRTL(p, experimental_data_array, experimental_gamma_array)
     
# call the optimizer -
opt_result = optimize(OF,xinitial,BFGS())
```

```julia
# build dictionary -
pVector = Optim.minimizer(opt_result)
model_parameters_dictionary = Dict{String,Any}()
model_parameters_dictionary["âº"] = pVector[1]
model_parameters_dictionary["a12"] = pVector[2]
model_parameters_dictionary["a21"] = pVector[3]
model_parameters_dictionary["b12"] = pVector[4]
model_parameters_dictionary["b21"] = pVector[5]

# compute the NRTL gamma values -
NRTL_gamma_array = compute_NRTL_gamma_array(experimental_data_array, model_parameters_dictionary)

# compute the composition using the NRTL model gamma's -
NRTL_composition_array = Array{Float64,2}(undef,number_of_temperature_steps,2)
for row_index = 1:number_of_temperature_steps
    
    # gamma -
    ð›„1 = NRTL_gamma_array[row_index,1]
    ð›„2 = NRTL_gamma_array[row_index,2]
    
     # get á´ª values -
    á´ª1 = psi_array[row_index,1]
    á´ª2 = psi_array[row_index,2]
    
    # compute -
    x1 = (1/ð›„1)*á´ª1
    x2 = 1 - x1
    
    NRTL_composition_array[row_index,1] = x1
    NRTL_composition_array[row_index,2] = x2 
end

plot(NRTL_composition_array[:,1],experimental_data_array[:,1],label="NRTL",lw=2)
plot!(ideal_composition_array[:,1],T_vector,label="Ideal",lw=2)
scatter!(experimental_data_array[:,2],experimental_data_array[:,1], label="Data Luo et al")
xlabel!("glutaric acid mol fraction x1 (A.U)")
ylabel!("Temperature T (K)")
```
